```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
```

```python
# --- Function to parse only selected keys ---
def parse_selected_scores(text, keys=("Google", "DoorDash", "TripAdvisor", "OpenTable", "Facebook")):
    if pd.isna(text):
        return {}
    pairs = [x.strip() for x in text.split(",")]
    result = {}
    for p in pairs:
        if ":" in p:
            key, val = p.split(":")
            key = key.strip()
            if key in keys:  # only keep selected keys
                result[key] = float(val)
    return result

# Initialize LabelEncoder
le = LabelEncoder()
```
